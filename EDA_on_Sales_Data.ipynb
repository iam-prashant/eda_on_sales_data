```python
sns.set(rc={'figure.figsize':(20,8)})
```

```python
a = sns.countplot(x='City',data=my_data)
plt.title("Orders from same city",fontsize=35)
plt.xlabel('City',fontsize=30)
plt.ylabel('Total Counts',fontsize=30)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.tight_layout()
plt.show()
```

```python
city_sales = my_data.groupby(['City'],as_index=False)['Sales'].sum().sort_values(by='Sales',ascending=False)
```

```python
city_sales
```

```python
sns.set(rc={'figure.figsize':(10,5)})

sns.barplot(data=city_sales,x='City',y='Sales')
plt.title("Total Sales from Different city",fontsize=20)
plt.xlabel('City',fontsize=13)
plt.ylabel('Sales',fontsize=13)
plt.xticks(fontsize=9,rotation=90,ha='left')
plt.yticks(fontsize=9)
plt.tight_layout()
plt.show()
```

```python
city_ordered_quantity = my_data.groupby(['City'],as_index=False)['Quantity Ordered'].sum().sort_values(by='Quantity Ordered',ascending=False)
```

```python
city_ordered_quantity
```

```python
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(data=city_ordered_quantity,x='City',y='Quantity Ordered')

plt.title("Quantity Ordered from Different City",fontsize=20)
plt.xlabel('City',fontsize=13)
plt.ylabel('Quantities Ordered',fontsize=13)
plt.xticks(fontsize=9,rotation=90,ha='left')
plt.yticks(fontsize=9)
plt.tight_layout()
plt.show()
```

```python
sns.set(rc={'figure.figsize':(25,12)})

sns.countplot(x='Product',data=my_data)
plt.xticks(rotation=45, ha='right',fontsize=20);
plt.yticks(fontsize=20);
plt.xlabel('Product',fontsize=25)
plt.ylabel('Count',fontsize=25)
plt.title('Product Count',fontsize=30)
plt.tight_layout()
plt.show()
```

```python
product_city_totals = my_data.groupby(['City', 'Product'])['Quantity Ordered'].sum().reset_index()

maxi = product_city_totals.groupby('City')['Quantity Ordered'].idxmax()
top_products_city = product_city_totals.loc[maxi]

plt.figure(figsize=(12, 8))
sns.barplot(data=top_products_city, x='City', y='Quantity Ordered', hue='Product')
plt.xlabel('City',fontsize=20)
plt.ylabel('Total Quantity Ordered',fontsize=20)
plt.title('Top Products Ordered in Each City',fontsize=30)
plt.xticks(fontsize=13);
plt.yticks(fontsize=13);
plt.tight_layout()

plt.show()
```

```python
product_city_totals = my_data.groupby(['City', 'Product'])['Quantity Ordered'].sum().reset_index()
sorted_product= product_city_totals.sort_values(by='Quantity Ordered', ascending=False)

plt.figure(figsize=(25, 10))
sns.barplot(data=sorted_product, x='City', y='Quantity Ordered', hue='Product',linewidth=2.5)
plt.xlabel('City',fontsize=30)
plt.ylabel('Total Quantity Ordered',fontsize=30)
plt.title('Total Products Ordered in Each City',fontsize=35)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left',fontsize=21)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.tight_layout()
plt.show()
```

```python
sorted_product_sales = my_data.groupby(['Product'],as_index=False)['Sales'].sum().sort_values(by='Sales',ascending=False)
sorted_product_sales
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sorted_product_sales = my_data.groupby(['Product'],as_index=False)['Sales'].sum().sort_values(by='Sales',ascending=False)

sns.barplot(data=sorted_product_sales, x='Product', y='Sales')
plt.xticks(rotation=45, ha='right',fontsize=20);
plt.yticks(fontsize=20);
plt.xlabel('Product',fontsize=25)
plt.ylabel('Total Sales',fontsize=25)
plt.title('Total Sales from Different Product',fontsize=30)
plt.tight_layout()
plt.show()
```
